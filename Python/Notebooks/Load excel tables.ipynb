```python
from openpyxl import load_workbook
import pandas as pd
import numpy as np
```

```python
input_file = "input_files/input_combined.xlsx"
```

```python
def table_to_df(ws, cell_range):
    cell_range = cell_range.split(":")
    data_rows = []
    for row in ws[cell_range[0]:cell_range[1]]:
        data_cols = []
        for cell in row:
            data_cols.append(cell.value)
        data_rows.append(data_cols)

    df = pd.DataFrame(data_rows)
    df.columns = df.loc[0]
    df.index = df[df.keys()[0]]
    
    df = (
            df
            .drop(df.keys()[0])
            .drop(columns=df.keys()[0])
            .replace("true", True)
            .replace("false", False)
            .replace({np.nan: None})
        )   
    
    return df

def read_tables(input_file):
    wb = load_workbook(filename = input_file)
    tables = {}
    for ws in wb.worksheets:
        for tbl in ws._tables:
            tables[tbl.name] = table_to_df(ws, tbl.ref)
    
    return tables
```

```python
read_tables(input_file)
```
