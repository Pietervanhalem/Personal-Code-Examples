```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import datetime

import numpy as np
from scipy.interpolate import interp1d as interpol


def sum_irregular_timeseries(x1, x2, y1, y2):
    new_x1 = list(x1)
    new_x2 = list(x2)

    for i in range(len(x1)):
        if type(x1[i]) == datetime.datetime:
            new_x1[i] = x1[i].timestamp()
    for i in range(len(x2)):
        if type(x2[i]) == datetime.datetime:
            new_x2[i] = x2[i].timestamp()

    int1 = interpol(new_x1, y1)
    int2 = interpol(new_x2, y2)

    new_x1 = np.array(new_x1)
    new_x2 = np.array(new_x2)

    if max(new_x1) > max(new_x2):
        new_x1 = new_x1[new_x1 <= max(new_x2)]
    elif max(new_x1) < max(new_x2):
        new_x2 = new_x2[new_x2 <= max(new_x1)]

    if min(new_x1) < min(new_x2):
        new_x1 = new_x1[new_x1 >= min(new_x2)]
    elif min(new_x1) > min(new_x2):
        new_x2 = new_x2[new_x2 >= min(new_x1)]

    new_x1 = list(new_x1)
    new_x2 = list(new_x2)
    new_x1.extend(new_x2)

    x = list(set(new_x1))
    x = list(np.sort(x))

    y = int1(x) + int2(x)
    return x, y
```

```python
x1 =np.array([2,3,5,10])
x2 =np.array([0,4,8,11])
y1 =np.array([1,2,3,0])
y2 =np.array([1,4,6,3])
x,y = sum_irregular_timeseries(x1,x2,y1,y2)
plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)
```
