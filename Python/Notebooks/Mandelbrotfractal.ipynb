```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import cv2
```

```python
X = 1000
Y = 1000
x_range = [-2,2]
y_range = [-2,2]
N = 100

pixels = np.ones((X,Y,3))
pixels[:,:,0] = 0
pixels[:,:,1] = 0
pixels[:,:,2] = 0


for i in range(X):
    for j in range(Y):
        y = i * (x_range[1] - x_range[0]) / X +  x_range[0]
        x = j * (y_range[1] - y_range[0]) / Y +  y_range[0]
        cx = x*1.0
        cy = y*1.0
        for n in range(N):
            xx = x**2 - y**2
            yy = 2*x*y
            x = xx+ cx
            y = yy+cy
            
            
            if abs(y+x) > 10:
                break
        
        if n < N-1:
            pixels[i,j,0] = 1 / (1 + abs(y)/10)
            pixels[i,j,1] = 1 / (1 + abs(x)/10)
            pixels[i,j,2] = 1 / (1+ abs(y+x)/10)
        else:
            pixels[i,j,0] = 0
            pixels[i,j,1] = 0
            pixels[i,j,2] = 0
            
fig = plt.figure(figsize=(15,15))            
plt.imshow(pixels)
```
