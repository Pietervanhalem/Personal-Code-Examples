```python
def FizzBuzz(
    i, 
    fizz_buzz = {
        'Fizz': 3, 
        "Buzz": 5, 
    }
):
    output = ''
    for key in fizz_buzz:
        output += key if i%int(fizz_buzz[key]) == 0 else ''
    output = str(i) if output == '' else output
    output = '0' if i == 0 else output
    return output
```

```python
for i in range(50,60):
    print(FizzBuzz(i))
```

```python
def PrimeFactorisation(i):
    factors = []
    rest = i
    j = 2
    while j< rest+1:
        if rest/j == int(rest/j):
            factors.append(j) if j != i else None
            rest = rest/j
        else:
            j = j + 1
    return sorted(factors)

import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
%%time

start_T = datetime.datetime.now().timestamp()
T = 0
CTS = []
j = 1
while T < 10*60:
    j = j + 1
    start = datetime.datetime.now()
    D = {}
    for i in range(j):
        D[i] = PrimeFactorisation(i)
    stop = datetime.datetime.now()
    CT = stop.timestamp() - start.timestamp()
    CTS.append(CT)
    T = datetime.datetime.now().timestamp() -  start_T
```

```python
CC = []
for i in range(len(D)):
    if len(D[i]) == 0:
        CC.append(True)
    else:
        CC.append(False)

CC = np.array(CC)[:-1]
CTS = np.array(CTS)
x = np.arange(len(CTS))
```

```python
plt.plot(x[CC == False], CTS[CC == False], 'b.', markersize = 1)
plt.plot(x[CC], CTS[CC], 'r.', markersize = 1)


plt.show()
```
